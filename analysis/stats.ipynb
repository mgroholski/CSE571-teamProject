```python
from scipy import stats
import subprocess
import matplotlib.pyplot as plt  # Common import for matplotlib
import numpy as np
import random
import os
from itertools import combinations
# Generating layouts
filenames=os.listdir("layouts")
val2=[]
laylay=[]
for f in filenames:
    val2 = (f.replace(".lay",""))
    laylay.append(val2)
float_listr=[]

randomLayout=random.choice(laylay)
numberOfEpisodes=80
trainEpisodes=60
command= [
        "python",
        "pacman.py",
        "-p", "ReinforceAgent",
        "-n", "{0}".format(numberOfEpisodes),
        "-x", "{0}".format(trainEpisodes),
        "-a", "alpha=0.2,gamma=0.8"
        "-q",
        "-l", "{0}".format(randomLayout)
    ]

result = subprocess.run(command, capture_output=True, text=True)
output=result.stdout

lines=output.splitlines()
value1=None
value2=[]
    
for line in lines:
    if 'Scores:' in line:
        value1 = (line.split(":")[1].strip())  # Extract integer value
        value2 = (value1.split(", "))
        float_list1=[float(value) for value in value2] 

###########################################        
###########################################
numberOfEpisodes=60
trainEpisodes=50

command= [
        "python",
        "pacman.py",
        "-p", "QLearningAgent",
        "-n", "{0}".format(numberOfEpisodes),
        "-x", "{0}".format(trainEpisodes),
        "-q",
        "-l", "{0}".format(randomLayout)
    ]

result = subprocess.run(command, capture_output=True, text=True)
output=result.stdout

lines=output.splitlines()
value1=None
value2=[]
    
for line in lines:
    if 'Scores:' in line:
        value1 = (line.split(":")[1].strip())  # Extract integer value
        value2 = (value1.split(", "))
        float_list2=[float(value) for value in value2] 
        
########################################        
########################################

numberOfEpisodes=80
trainEpisodes=60
command= [
        "python",
        "pacman.py",
        "-p", "ActorCriticAgent",
        "-n", "{0}".format(numberOfEpisodes),
        "-x", "{0}".format(trainEpisodes),
        "-a", "-a alpha_theta=0.25,alpha_w=0.15,gamma=0.9"
        "-q",
        "-l", "{0}".format(randomLayout)
    ]

result = subprocess.run(command, capture_output=True, text=True)
output=result.stdout

lines=output.splitlines()
value1=None
value2=[]
    
for line in lines:
    if 'Scores:' in line:
        value1 = (line.split(":")[1].strip())  # Extract integer value
        value2 = (value1.split(", "))
        float_list3=[float(value) for value in value2] 
        #float_list3.append(float_list1)

##################################
#Plot the episodes
i=np.arange(len(float_list1))+1
plt.plot(i,float_list1,'r',label="Reinforcement")
plt.plot(i,float_list2,'k',label="QLearningAgent")
plt.plot(i,float_list3,'b',label="ActorCriticAgent")

plt.title('Learning Methods vs. Training Numbers')
plt.xticks(i)
plt.xlabel('Training Episode')
plt.ylabel('Score per Episode')
plt.legend()
plt.savefig('learning_methods.png')
plt.show()

#################################
# T-test Statistics
Methods = {
    'Reinforcement': float_list1,
    'QLearning': float_list2,
    'Actor-Critic': float_list3
}
def finalttest(data_dict, alpha=0.05):
    # Get all unique pairs of keys and their corresponding lists
    key_pairs = list(combinations(data_dict.keys(), 2))
    
  
    results = {}
    
    # Perform t-tests for each pair
    for key1, key2 in key_pairs:
        t_stat, p_value = stats.ttest_ind(data_dict[key1], data_dict[key2])
        
        pair_name = f"{key1} vs {key2}"
        results[pair_name] = {
            "t_stat": t_stat,
            "p_value": p_value,
            "significant": p_value < alpha,
            "message": f'There is {"a significant" if p_value < alpha else "no significant"} difference between the means of {key1} and {key2}'
        }

    return results


results = finalttest(Methods)

for key, value in results.items():
    print(key)
    print("T-statistic:", value["t_stat"])
    print("P-value:", value["p_value"])
    print(value["message"])
    print("Significant:", value["significant"])
    print()
```
